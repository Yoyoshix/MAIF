```python
import numpy

good = "anyway✔bone✔cinema✔cover✔define✔economy✔eleven✔eye✔hesitate✔key✔majority✔music✔palace✔potato✔relax✔royal✔surely✔typical✔vocabulary✔water"
bad  = "action✕blue✕certain✕crown✕equal✕gamble✕jacket✕lawyer✕multiply✕mysterious✕news✕odd✕opposite✕prize✕quit✕symbol✕thirteen✕toe✕try✕young"

good = good.split("✔")
bad = bad.split("✕")

for i in range(len(good)):
    good[i] = "^" + good[i] + "$"
for i in range(len(bad)):
    bad[i] = "^" + bad[i] + "$"

mapped_char = []

cpt = 0
for source in [good, bad]:
    for i in source:
        for j in i:
            if j not in mapped_char:
                mapped_char.append(j)

print(mapped_char)
nitem = len(mapped_char)
limit = 4
maxmax = 0
chain = numpy.zeros((limit,nitem,nitem))

for space in range(limit):
    for txt in good:
        copycat = []
        for i in range(len(txt)-space-1):
            idx1 = mapped_char.index(txt[i])
            idx2 = mapped_char.index(txt[i+space+1])
            if idx1*nitem+idx2 not in copycat:
                chain[space][idx1][idx2] += 1
            copycat.append(idx1*nitem+idx2)
            if maxmax < chain[space][idx1][idx2]:
                maxmax = chain[space][idx1][idx2]
            
for space in range(limit):
    for txt in bad:
        for i in range(len(txt)-space-1):
            idx1 = mapped_char.index(txt[i])
            idx2 = mapped_char.index(txt[i+space+1])
            chain[space][idx1][idx2] = 0

ranker = [0] * int(maxmax)
for space in range(len(chain)):
    for i in range(len(chain[0])):
        for j in range(len(chain[0][0])):
            ranker[int(chain[space][i][j])] += 1

for score in range(len(ranker), 0, -1):
    for space in range(len(chain)):
        for i in range(len(chain[0])):
            for j in range(len(chain[0][0])):
                if chain[space][i][j]-space == score:
                    print(score, mapped_char[i], mapped_char[j], space)
```
