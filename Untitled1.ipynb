```python
# Formula based Python3 code to find  
# sum of all divisors of n. 
import math as m
import time

# Returns sum of all factors of n. 
def sumofFactors(n):
    # Traversing through all 
    # prime factors
    res = 1
    curr_sum = 1
    curr_term = 1
    while n % 2 == 0:
        n = n // 2
        curr_term = curr_term * 2
        curr_sum += curr_term
    res = res * curr_sum 
    
    for i in range(3, int(n**0.5)+1, 2):
        curr_sum = 1
        curr_term = 1
        while n % i == 0:
            n = n // i
            curr_term = curr_term * i
            curr_sum += curr_term
        res = res * curr_sum
    
    # This condition is to handle the  
    # case when n is a prime number  
    # greater than 2 
    if n > 2:
        res = res * (1 + n) 
    return res

def get_sum_div(n):
    res = 0
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            res += i
            res += n // i
    if i**2 == n:
        res -= i
    return res

def prime_factor(n):
    prime_factor_list = [] #I'm not using dictionnary cause I need to access values by id later
    prime_factor_amount = []
    div_amount = 1
    length = 0
    if n % 2 == 0:
        length += 1
        prime_factor_list.append(2)
        prime_factor_amount.append(0)
        while n % 2 == 0:
            prime_factor_amount[0] += 1
            n = n // 2
        div_amount *= prime_factor_amount[0] + 1
    for div in range(3, int(n**0.5)+1, 2):
        if n % div == 0:
            length += 1
            prime_factor_list.append(div)
            prime_factor_amount.append(0)
            while n % div == 0:
                prime_factor_amount[-1] += 1
                n = n // div
            div_amount *= prime_factor_amount[-1] + 1
    if n > 1:
        length += 1
        prime_factor_list.append(n)
        prime_factor_amount.append(1)
        div_amount *= 2
    return prime_factor_list, prime_factor_amount, div_amount, length

def get_sum_div_prime(n):
    prime_factor_list, prime_factor_amount, total_div, length = prime_factor(n)
    
    res = 0
    table_count = [0] * length
    cursor = 0
    div = 1
    current_div = 1
    while current_div < total_div:
        res += div + n // div
        table_count[cursor] += 1
        if table_count[cursor] > prime_factor_amount[cursor]:
            while table_count[cursor] > prime_factor_amount[cursor]:
                table_count[cursor] = 0
                div = div // (prime_factor_list[cursor] ** prime_factor_amount[cursor])
                cursor += 1
                table_count[cursor] += 1
            div *= prime_factor_list[cursor]
            cursor = 0
        else:
            div *= prime_factor_list[cursor]
        current_div += 2
    if total_div % 2 == 1:
        res += div
    return res

n = 60
s1 = time.time()
sum1 = sumofFactors(n)
e1 = time.time()

s2 = time.time()
sum2 = get_sum_div(n)
e2 = time.time()

s3 = time.time()
sum3 = get_sum_div_prime(n)
e3 = time.time()
print(n, sum1, e1 - s1, e2 - s2, e3 - s3)
if len(set([sum1, sum2, sum3])) != 1:
    print(n, sum1, sum2, sum3)
```

```python
def sieveOfEratosthenes(N, s): 
      
    # Create a boolean array  
    # "prime[0..n]" and initialize 
    # all entries in it as false. 
    prime = [False] * (N+1) 
  
    # Initializing smallest factor 
    # equal to 2 for all the even  
    # numbers 
    for i in range(2, N+1, 2):  
        s[i] = 2
  
    # For odd numbers less then  
    # equal to n 
    for i in range(3, N+1, 2): 
        if (prime[i] == False): 
              
            # s(i) for a prime is 
            # the number itself 
            s[i] = i 
  
            # For all multiples of 
            # current prime number 
            for j in range(i, N // i + 1, 2): 
                if (prime[i*j] == False): 
                    prime[i*j] = True

                    # i is the smallest  
                    # prime factor for 
                    # number "i*j". 
                    s[i * j] = i

# factors and its power 
def generatePrimeFactors(N): 
    # s[i] is going to store
    # smallest prime factor
    # of i.
    s = [0] * (N+1)
  
    # Filling values in s[]
    # using sieve
    sieveOfEratosthenes(N, s)
    print(s)
  
    print("Factor Power")
  
    # Current prime factor of N
    curr = s[N]
      
    # Power of current prime factor
    cnt = 1
  
    # Printing prime factors and
    #their powers
    while (N > 1):
        N //= s[N]
  
        # N is now N/s[N]. If new N
        # als has smallest prime
        # factor as curr, increment
        # power
        if (curr == s[N]):
            cnt += 1
            continue
        print(str(curr) + "\t" + str(cnt))
  
        # Update current prime factor
        # as s[N] and initializing
        # count as 1.
        curr = s[N]
        cnt = 1

N = 360
generatePrimeFactors(N)
```
