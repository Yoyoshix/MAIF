```python
import numpy as np
import copy

class Minemap():
    def __init__(self, hei, wid, mine, radius=5, density=3, safe_start=True):
        self.hei = hei
        self.wid = wid
        self.mine = mine
        self.density = density
        self.radius = radius
        self.safe_start = safe_start
        
        self.target_x = 0
        self.target_y = 0
        self.cell_map = []
        self.luck_map = []

    def get_neighbour(self, x, y, value):
        res = 0
        for i in range( -(x != 0), 1 + (x != self.hei - 1)):
            for j in range( -(y != 0), 1 + (y != self.wid - 1)):
                res += (self.cell_map[x + i][y + j] == value)
        return (res)

    def travel(self, cell_map, x, y, start_x, start_y):
        pass

    def is_placeable(self, cell_map, x, y, hei, wid):
        neigh_bomb += (x == 0) + (x == hei-1) + (y == 0) + (y == wid-1)
        if x > 0:
            neigh_bomb += (cell_map[x-1][y] == -1)
        if x < hei-1:
            neigh_bomb += (cell_map[x+1][y] == -1)
        if y > 0:
            neigh_bomb += (cell_map[x][y-1] == -1)
        if y < wid-1:
            neigh_bomb += (cell_map[x][y+1] == -1)
        if neigh_bomb <= 1:
            return 1
        cell_map[x][y] = -1



        return 1

    def get_distance(self, x, y, pos_x, pos_y):
        return ((x-pos_x)**2+(y-pos_y)**2)**0.5

    def lower_lucky(self, x, y):
        min_x = (x-self.radius) * (x-self.radius > 0)
        min_y = (y-self.radius) * (y-self.radius > 0)
        max_x = self.hei - (self.hei-(x+self.radius)-1) * (x+self.radius < self.hei)
        max_y = self.wid - (self.wid-(y+self.radius)-1) * (y+self.radius < self.wid)
        #print(min_x, min_y, max_x, max_y)
        for i in range(min_x, max_x):
            for j in range(min_y, max_y):
                incr = self.get_distance(x, y, i, j)-1
                if incr < self.radius:
                    self.luck_map[i][j] = (1 - self.luck_map[i][j]) * incr

    def generate_map(self):
        empty_cell = self.hei * self.wid
        self.cell_map = np.zeros((self.hei, self.wid))
        self.luck_map = np.zeros((self.hei, self.wid))
        nb = [i for i in range(empty_cell)]
        stop = empty_cell
        
        while empty_cell+self.mine > stop:
            rdm = nb[np.random.randint(0, empty_cell)] #random between 0 and amount of empty cell to get an empty cell value
            x = rdm // self.wid #get column
            y = rdm % self.wid #get line
            if np.random.random() > self.luck_map[x][y]:# and is_placeable(copy.copy(cell_map), x, y, hei, wid) == 1:
                self.cell_map[x][y] = -1 #put mine at coordinates
                empty_cell -= 1
                nb[rdm] = nb[empty_cell] #swap the targeted cell (which is now a mine) with the last element of the array
                self.lower_lucky(x, y)
            else:
                pass
                #print("cancelled")

        for i in range(self.hei):
            for j in range(self.wid):
                if self.cell_map[i][j] != -1:
                    self.cell_map[i][j] = self.get_neighbour(i, j, -1)
        #print(self.luck_map)

    def print_map(self):
        for i in self.cell_map:
            line = ""
            for j in i:
                if j == -1:
                    line += "X"
                if j >= 0:
                    line += " "
            line += "|"
            for j in i:
                if j == -1:
                    line += "X"
                if j == 0:
                    line += " "
                if j > 0:
                    line += str(int(j))
            line += "|"
            print(line)
        print("-"*len(self.cell_map))

    
    def check(self):
        count = 0
        for i in self.cell_map:
            for j in i:
                count += (j == -1)
        if count != self.mine:
            self.print_map()
            return -1
        return 0
```
